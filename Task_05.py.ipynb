```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
y = np.array([23, 56, 41, 62, 19] )

plt.subplot(2, 2, 1)
plt.bar(x,y)

x = np.random.normal([12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10] )
plt.subplot(2, 2, 2)
plt.hist(x)


y = np.array([30, 25, 15, 30])
mylabels = ['Apple', 'Banana', 'Orange', 'Mango'] 
plt.subplot(2, 2, 3)
plt.pie(y, labels = mylabels)

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5, 3, 6, 4])

plt.subplot(2, 2, 4)
plt.scatter(x,y)
```
